```python
import os

def rename_files_in_folders(folder_a, folder_b, name_mapping):
    # 遍历A文件夹中的文件
    print(f'Renaming files in {folder_a}...')
    for filename in os.listdir(folder_a):
        if filename in name_mapping:
            old_file_path = os.path.join(folder_a, filename)
            new_file_path = os.path.join(folder_a, name_mapping[filename])
            os.rename(old_file_path, new_file_path)
            print(f'Renamed: {old_file_path} -> {new_file_path}')

    # 遍历B文件夹中的文件
    print(f'Renaming files in {folder_b}...')
    for filename in os.listdir(folder_b):
        if filename in name_mapping:
            old_file_path = os.path.join(folder_b, filename)
            new_file_path = os.path.join(folder_b, name_mapping[filename])
            os.rename(old_file_path, new_file_path)
            print(f'Renamed: {old_file_path} -> {new_file_path}')
```

```python
filenames = os.listdir('./generated_gray_image/')
filenames.sort()
format_name = [str(i)+'.png' for i in range(1,153)]
filename_mapping = {original: mapped for original, mapped in zip(filenames, format_name)}

cache_1 = './generated_gray_image/'
cache_2 = './targeted_gray_image/'
rename_files_in_folders(cache_1, cache_2, filename_mapping)
```

```python
import os
import shutil

# 定义源文件夹和目标文件夹
data_folder = './generated_gray_image'
train_folder = os.path.join(data_folder, 'train')
val_folder = os.path.join(data_folder, 'val')
test_folder = os.path.join(data_folder, 'test')

# 创建目标文件夹
os.makedirs(train_folder, exist_ok=True)
os.makedirs(val_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# 遍历data文件夹中的所有文件
for filename in os.listdir(data_folder):
    if filename.endswith('.png'):
        # 获取文件名中的数字部分
        file_number = int(filename.split('.')[0])
        
        # 根据文件编号决定目标文件夹
        if 1 <= file_number <= 120:
            target_folder = train_folder
        elif 121 <= file_number <= 140:
            target_folder = val_folder
        elif 141 <= file_number <= 152:
            target_folder = test_folder
        else:
            continue  # 忽略不在范围内的文件
        
        # 构建源文件路径和目标文件路径
        source_path = os.path.join(data_folder, filename)
        target_path = os.path.join(target_folder, filename)
        
        # 移动文件
        shutil.move(source_path, target_path)

print("文件移动完成！")
```
